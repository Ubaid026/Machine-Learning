```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as st
```

```python
df.drop(columns=["Reviews"],inplace=True)
```

```python
df["Rating (Out of 5)"].mode()
```

```python
df["Rating (Out of 5)"]=df["Rating (Out of 5)"].astype("float")
```

```python
df["Rating (Out of 5)"].agg(["mean","median"])
```

```python
df["Rating (Out of 5)"].fillna(df["Rating (Out of 5)"].median(),inplace=True)
```

```python
df["Selling Price"]=df["Selling Price"].str.replace(",","").astype("float")
```

```python
df["Original Price"]=df["Original Price"].str.replace(",","").astype("int")
```

```python
df["Average Battery Life (in days)"]=df["Average Battery Life (in days)"].astype("int")
```

```python
df["discount"]=df["Original Price"]-df["Selling Price"]
df
```

```python
df["Discount Percentage"]=round((df["discount"]/df["Selling Price"])*100,2)
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df1=df.select_dtypes("object")
df1
```

```python
df1.drop(columns=["Model Name","Color"],inplace=True)
```

```python
df1.columns
```

```python
lbe=LabelEncoder()

df["Brand Name"]=lbe.fit_transform(df1["Brand Name"])
df["Device Type"]=lbe.fit_transform(df1["Device Type"])
df["Display"]=lbe.fit_transform(df1["Display"])
```

```python
df.drop(columns=["Model Name","Color","Strap Material"],inplace=True)
```

```python
df
```

```python
x=df.drop(columns=["Selling Price"])
y=df["Selling Price"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.03,random_state=123)
```

```python
from sklearn.ensemble import AdaBoostRegressor,GradientBoostingRegressor
import
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,y_pred)
```
